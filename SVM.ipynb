### Análise Variável de Composição, Saída vs Entrada.

Carbono (C)

C vs CO = +0.074
Pequena correlação positiva. Isso sugere que o teor de carbono da biomassa não influencia fortemente a geração de CO diretamente, mas pode contribuir indiretamente via reações com CO₂ (Boudouard) ou H₂O (água-gás).

C vs CO₂ = −0.148
Biomassas mais carbonosas tendem a produzir menos CO₂. Isso reforça a formação de CO (via Boudouard) em vez de CO₂.

C vs CH₄ = −0.004
Essencialmente nula. O teor de carbono não está correlacionado diretamente com CH₄.

C vs H₂ = +0.378
Correlação significativa. Maior teor de carbono → mais reações água-gás e Boudouard → mais hidrogênio como subproduto.

Hidrogênio (H)

H vs CO = +0.086
Pequena correlação positiva, sem grande significado físico direto.

H vs CO₂ = +0.248
Biomassas mais ricas em H tendem a gerar mais CO₂, possivelmente pela oxidação de H em H₂O, seguida por reações deslocamento água-gás.

H vs CH₄ = +0.131
Correlação moderada → presença de mais hidrogênio favorece a formação de CH₄ (reação de metanação).

H vs H₂ = +0.182
Esperada, porém mais fraca do que o esperado. Pode indicar que a formação de H₂ depende mais da reação com vapor (Steam/Biomass) do que do teor de H original na biomassa.

Oxigênio (O)

O vs CO = −0.147
Biomassas mais oxigenadas produzem menos CO, provavelmente devido à oxidação do carbono diretamente para CO₂.

O vs CO₂ = +0.056
Fraca correlação positiva. Reflete a tendência de biomassa rica em O formar mais CO₂ (oxidação direta).

O vs CH₄ = −0.031
Praticamente nula. O oxigênio da biomassa não influencia diretamente o metano.

O vs H₂ = −0.401
Correlação negativa significativa. Maior oxigênio → menor H₂, possivelmente porque há mais formação de H₂O e CO₂, reduzindo o H₂ livre.

Nitrogênio (N)

N vs CO = +0.086
Fraca e sem relevância técnico-química direta.

N vs CO₂ = −0.156
Pode indicar que biomassa com mais N tende a oxidar menos carbono em CO₂, ou efeito indireto da composição.

N vs CH₄ = −0.139
Pequeno impacto negativo. Nitrogênio não contribui para formação de CH₄ e pode indicar menor proporção de compostos voláteis formadores de hidrocarbonetos.

N vs H₂ = −0.304
Correlação negativa. Biomassas mais nitrogenadas tendem a formar menos H₂, possivelmente porque o N não participa das reações de geração de gás combustível e ocupa espaço na estrutura da biomassa que poderia gerar H.

#**SMOGN**

Como observado nas análises de estatística descritiva (resumo estatístico, boxplot e histograma), os dados apresentam alta variabilidade e desbalanceamento entre classes. Para mitigar esse problema,  foi utilizado o método SMOGN (Synthetic Minority Over-sampling Technique for Regression with Gaussian Noise) com o objetivo de equilibrar a distribuição de variáveis alvo raras relacionadas à composição do gás de síntese gerado na gaseificação de biomassa, além disso, todos os dados serão normalizados, garantindo a uniformidade de escala entre as variáveis.

Parâmetros Utilizados
Na função  apply_smogn_multitarget, foram definidos os seguintes parâmetros principais:

rel_thres: Define o limiar relativo para identificar amostras raras.

k: Número de vizinhos utilizados na interpolação para geração sintética.

rel_method = "auto": Método automático de definição do critério de raridade com base nas distribuições individuais de cada alvo.

under_smpl = False: Não foi aplicado undersampling, preservando a distribuição original das classes majoritárias.


```python
!pip install imblearn smogn scikit-learn
import smogn
from sklearn.preprocessing import RobustScaler
```

```python
def apply_smogn_multitarget(df, targets, max_frac=0.4, rel_thres=0.9, k=2, random_state=42, rel_ctrl_pts_rg=None):
    """
    Aplica SMOGN para balanceamento de dados multirrótulo.

    Parâmetros:
        df: DataFrame com os dados originais.
        targets: Lista de variáveis de saída para balanceamento.
        max_frac: Fração máxima de amostras sintéticas em relação ao original.
        rel_thres: Limiar de relevância para SMOGN automático.
        k: Número de vizinhos para SMOGN.
        rel_ctrl_pts_rg: Pontos de controle para função de relevância manual (opcional).

    Retorna:
        DataFrame balanceado com originais + sintéticas.
    """
    import pandas as pd
    import numpy as np
    from sklearn.preprocessing import RobustScaler
    import smogn

    df_num = df.select_dtypes(include='number').copy()

    for target in targets:
        if target not in df_num.columns:
            raise ValueError(f"Variável de saída '{target}' não está no DataFrame ou não é numérica.")

    n_original = len(df)
    max_rows = int(n_original * (1 + max_frac))
    synthetic_dfs = []
    scaler = RobustScaler()
    df_scaled = pd.DataFrame(scaler.fit_transform(df_num), columns=df_num.columns)

    generated_per_target = {target: 0 for target in targets}

    for y in targets:
        try:
            smogn_params = {
                'data': df_scaled,
                'y': y,
                'k': k,
                'samp_method': 'balance',
                'drop_na_col': True,
                'drop_na_row': True,
                'replace': False,
                'pert': 0.04
            }

            if y == 'CO2 [%vol]' and rel_ctrl_pts_rg is not None:
                smogn_params.update({
                    'rel_method': 'manual',
                    'rel_ctrl_pts_rg': rel_ctrl_pts_rg
                })
            else:
                smogn_params.update({
                    'rel_method': 'auto',
                    'rel_thres': rel_thres
                })

            df_synth = smogn.smoter(**smogn_params)

            df_synth_tuples = df_synth.apply(tuple, axis=1)
            df_scaled_tuples = df_scaled.apply(tuple, axis=1)
            new_samples_mask = ~df_synth_tuples.isin(df_scaled_tuples)
            df_extra_scaled = df_synth[new_samples_mask].dropna().drop_duplicates()

            if not df_extra_scaled.empty:
                df_extra = pd.DataFrame(
                    scaler.inverse_transform(df_extra_scaled),
                    columns=df_num.columns
                ).drop_duplicates()

                generated_per_target[y] = len(df_extra)
                synthetic_dfs.append(df_extra)

        except Exception as e:
            continue

    if not synthetic_dfs:
        print("\n❌ Nenhuma amostra sintética gerada.")
        return df_num

    df_synthetic = pd.concat(synthetic_dfs, ignore_index=True).drop_duplicates()
    num_synthetic_needed = max_rows - n_original

    if len(df_synthetic) > num_synthetic_needed and num_synthetic_needed > 0:
        df_synthetic_final = df_synthetic.sample(n=num_synthetic_needed, random_state=random_state)
    elif num_synthetic_needed <= 0:
        df_synthetic_final = pd.DataFrame(columns=df_num.columns)
    else:
        df_synthetic_final = df_synthetic

    final_df = pd.concat([df_num, df_synthetic_final], ignore_index=True)

    # ✅ Relatório final SIMPLIFICADO: apenas total de amostras sintéticas
    total_synthetic = len(df_synthetic_final)
    print(f"\n✅ Total de amostras sintéticas geradas: {total_synthetic}")

    return final_df
```

```python
df_train = pd.concat([X_train, y_train], axis=1)
```

# **Construção do Modelo SVM (SVR) - Dados Sim SMOGN**

# ***METODOLOGIA***

Para o desenvolvimento do modelo de Support Vector Regression (SVR), foram previamente aplicadas as etapas de preparação dos dados e treinamento da base de dados. Esse modelo faz parte da família de algoritmos de Máquinas de Vetores de Suporte (SVM), que utilizam núcleos lineares e não lineares para encontrar um hiperplano ótimo que se ajuste aos dados.

Com o objetivo de melhorar o ajuste do modelo, foram definidas restrições na função Kernel, que transforma as variáveis de entrada 𝑥 e saída 𝑦. Utilizou-se a função núcleo RBF (função Base Radial), que podem ser escritas na forma ***K (x,y)= f(d(x,y))***, onde d e uma metrica em X e F e uma funçao em R+.Usualmente a metrica vem do produto interno ***d(x,y)=∥𝑥−𝑦∥= (x-y,x-y) exp1/2***. Então podemos definir a ecuação:

**𝐾(𝑥,𝑦)=exp(⁡−𝛾∥𝑥−𝑦∥2)**     Ecuação (10)

Para definir o hiperplano que melhor se ajusta aos dados, foram estabelecidas algumas restrições na função Kernel, como:

*   Gamma (γ): Influencia a forma da região tubular gerada pela função RBF, afetando diretamente a complexidade do modelo, (gamma 'auto').
*   Epsilon (ε): Estabelece a margem de tolerância para o erro. Neste caso, foi fixado em 0.5.
* C(penalização): Regula o grau de ajuste do modelo. Um valor alto de 𝐶 penaliza fortemente os erros (potencial de overfitting), enquanto um valor baixo permite mais erros. Foi adotado 𝐶=100 para buscar um equilíbrio adequado.

Essas restrições permitem o cálculo dos produtos internos entre os vetores de suporte, que são então combinados com os pesos do modelo para gerar a predição final.
